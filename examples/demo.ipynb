# DuckDB Jupyter Notebook Demo

Welcome! Each code cell below executes SQL against DuckDB using the DuckDB SQL kernel.

```python
-- Check DuckDB version
SELECT version() AS duckdb_version;
```

```python
-- Create a sample table
CREATE TABLE users (
    id INTEGER PRIMARY KEY,
    name VARCHAR,
    age INTEGER,
    city VARCHAR
);

INSERT INTO users VALUES
    (1, 'Alice', 30, 'New York'),
    (2, 'Bob', 25, 'San Francisco'),
    (3, 'Charlie', 35, 'Seattle'),
    (4, 'Diana', 28, 'New York'),
    (5, 'Eve', 32, 'San Francisco');
```

```python
-- Query the data
SELECT * FROM users ORDER BY age DESC;
```

```python
-- Aggregation
SELECT 
    city,
    count(*) AS user_count,
    avg(age) AS avg_age,
    min(age) AS youngest,
    max(age) AS oldest
FROM users
GROUP BY city
ORDER BY user_count DESC;
```

## Window Functions

DuckDB supports advanced SQL features like window functions.

```python
-- Window function example
SELECT 
    name,
    city,
    age,
    ROW_NUMBER() OVER (PARTITION BY city ORDER BY age DESC) AS age_rank_in_city,
    AVG(age) OVER (PARTITION BY city) AS city_avg_age
FROM users
ORDER BY city, age_rank_in_city;
```
